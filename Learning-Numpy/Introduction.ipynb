## What is Numpy
Numpy is a python package which runs on C under the hood and it is a very fast package which allows us to do computing and calculations. it contains N Dimensional Arrays which are very fast than Python Lists. Its full form is numerical python. 

## N Dimensional Arrays
These are arrays which are different as compared python lists these are homogenous i.e they contain similar type of data unlike lists which can contain hybrid type of data. Unlike lists, Numpy arrays are optimized with C programming language making it way faster than lists. Also they have contiguous memory. They support vectorized operations and have inbuilt mathematical functions. It has reduced overhead.
