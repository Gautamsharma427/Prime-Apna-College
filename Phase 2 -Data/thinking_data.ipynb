```python
import json
```

```python
# Loading Data from JSON files
def load_data(filename):
    with open(filename,"r") as file:
        data = json.load(file)
        return data
```

```python
data = load_data("data.json")
print(data,type(data))
```

```python
# data here is inconsistent(missing someplaces, folating data,multiple datatypes etc.)
# there is duplicate data too(assuming customer name is a unique value) 
```

```python
# cleaning the data

def cleaned_data(data):
    rating_map = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_daa = []
    unique_users = set()
    for user in data:
        # print(user['rating'])
        if user['rating'] in rating_map:
            user['rating'] = rating_map[user['rating']]

        #handling missing values
        raw_name = user.get('name')
        raw_rating = user.get('rating')
        raw_feedback = user.get('feedback')
        raw_age  = user.get('age')
        if raw_name == None:
            user['name'] = raw_name
        elif raw_rating == None:
            user['rating'] = raw_rating
        elif raw_feedback == None:
            user['feedback']=raw_feedback
        elif raw_age == None:
            user['age'] = raw_age

        #deduplication
        if user['name'] in unique_users:
            continue
        cleaned_daa.append(user)
        unique_users.add(user['name'])
    return cleaned_daa
        
```

```python
def dumpdata(data,filename):
    with open(filename, 'w') as file:
        json.dump(data,file,indent=4)
```

```python
data = cleaned_data(data)
```

```python
dumpdata(data,"cleanData.json")
```

```python
def get_insights(data):
    rate = []
    for ratings in data:
        #print(ratings['rating'])
        try:
            var = float(ratings['rating'].strip())
            rate.append(var)
        except Exception as e:
            rate.append(ratings['rating'])
            print("not executed",e)
    print(rate)
    sum = 0
    for value in rate:
        sum = sum+value
    average = sum/len(rate)
    print("average is:",average)

    poor_ratings = 0 
    total_ratings = len(rate)
    for rating in rate:
        if rating<4.0:
            poor_ratings+=1
    print(poor_ratings)

    print(f"%age of Poor Ratings is: {poor_ratings/total_ratings * 100}")
```

```python
def recommendation_system(data):
    recommendations = []
    for user in data:
        current_recommendation = {}
        current_recommendation['name'] = user['name']
        if(float(user['rating'])>3):
            current_recommendation['brand']="apple"
        else:
            current_recommendation['brand'] = "samsung"
        recommendations.append(current_recommendation)
    return recommendations
    
```

```python
with open('cleandata.json',"r") as f:
    data = json.load(f)
    get_insights(data)
    print(recommendation_system(data))
```
